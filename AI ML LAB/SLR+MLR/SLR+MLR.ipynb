```python
import pandas as pd


data = pd.read_csv("Advertising.csv")

# Display the first few rows of the dataset to understand its structure
data.head()
```

```python

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
# Create a figure and a set of subplots
fig, ax = plt.subplots()
# Create a boxplot
ax.boxplot([data['TV'], data['Newspaper'], data['Sales']])
# Set the labels for the boxplot
ax.set_xticklabels(['TV', 'Newspaper' , 'Sales'])
# Display the plot
plt.tight_layout()
```

```python
#scatter plot for relation between tv and sales
plt.scatter(data['TV'], data['Sales'], color='blue')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.title('Scatter Plot - TV Advertising vs Sales')
plt.show()

```

```python
#SLR

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Independent variable (TV) and dependent variable (Sales)
X = data[['TV']]
Y = data['Sales']

# Split the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create and train the model
model = LinearRegression()
model.fit(X_train, Y_train)

# Predict using the test data
Y_pred = model.predict(X_test)

# Visualizing the regression line and data points
plt.scatter(X_test, Y_test, color='blue', label='Actual Sales')
plt.plot(X_test, Y_pred, color='red', label='Predicted Sales')

# Adding labels and title
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.title('Simple Linear Regression - Sales vs TV Advertising')
plt.legend()

# Show plot
plt.show()

# Coefficient and Intercept of the linear regression line
slope = model.coef_[0]
intercept = model.intercept_

slope, intercept
```

```python
#MLR

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np

# Independent variables (TV, Radio, Newspaper) and dependent variable (Sales)
X = data[['TV', 'Radio', 'Newspaper']]
Y = data['Sales']

# Split the dataset into training and testing sets (80% train, 20% test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create and train the model
mlr_model = LinearRegression()
mlr_model.fit(X_train, Y_train)

# Predict using the test data
Y_pred = mlr_model.predict(X_test)

# Plotting Actual vs Predicted Sales
plt.scatter(Y_test, Y_pred, color='blue', label='Predicted vs Actual')

# Plotting the line of equality (ideal case)
plt.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], color='red', linestyle='--', label='Ideal Line')

# Adding labels and title
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.title('Actual vs Predicted Sales - Multiple Linear Regression')
plt.legend()

# Show plot
plt.show()

# Coefficients and Intercept
print(f"Coefficients (TV, Radio, Newspaper): {mlr_model.coef_}")
print(f"Intercept: {mlr_model.intercept_}")
```
