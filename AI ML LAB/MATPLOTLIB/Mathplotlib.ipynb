```python
# Data Visualization with Matplotlib

# Line Chart

import matplotlib.pyplot as plt
years = [2010, 2011, 2012, 2013, 2014, 2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(years, yield_apples)
plt.xlabel("Year")
plt.ylabel("Yield (tons per hectare)")
plt.title("Apple Yield Over Time")
plt.show()
```

```python
#Multiple Lines

apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.934, 0.936, 0.937]
oranges = [0.962, 0.941, 0.930, 0.918, 0.908, 0.907, 0.904, 0.901]
years = range(2000, 2008)
plt.plot(years, apples, marker='o', label="Apples")
plt.plot(years, oranges, marker='x', linestyle='--', label="Oranges")
plt.xlabel("Year")
plt.ylabel("Yield")
plt.title("Crop Yields in Kanto")
plt.legend()
plt.show()

```

```python
#Bar Plot

x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]
plt.bar(x, y)
plt.title("Bar Chart")
plt.show()

```

```python
#Scatter Plot

plt.scatter(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Chart")
plt.show()
```

```python
#Pie Chart

data = [1, 2, 3, 4]
explode = (0.1, 0, 0, 0)
plt.pie(data, explode=explode)
plt.title("Pie Chart")
plt.show()
```
